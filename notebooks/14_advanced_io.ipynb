## SQL Programming via Python

```python
# Example

# QUESTION: Compute the total qty product sold per product sorted by highest to lowest qty

# SQL version
conn = sqlite3.connect("../data/northwind.db") # ensure this is the correct path for the sqlite file. 
query_df = pd.read_sql_query("select ProductID,sum(Quantity) as qty_sold from 'Order Details' group by ProductID order by qty_sold desc", conn)
 
display(query_df)
conn.close()
```

```python
import duckdb

# Connect to DuckDB
conn = duckdb.connect()

# Install and load the SQLite scanner extension
conn.execute("INSTALL sqlite_scanner;")  # Ensure the SQLite extension is installed
conn.execute("LOAD sqlite_scanner;")

# Attach the SQLite database
conn.execute("ATTACH DATABASE '../data/northwind.db' (FORMAT 'sqlite');")

# List the tables in the SQLite database
tables = conn.execute("SELECT * FROM sqlite_master WHERE type='table';").fetchdf()
print("Tables in SQLite database:")
display(tables)

```

```python
# Query the 'Order Details' table using the fully qualified name
query_df = conn.execute("""
    SELECT ProductID, SUM(Quantity) AS qty_sold
    FROM northwind."Order Details"
    GROUP BY ProductID
    ORDER BY qty_sold DESC
""").fetchdf()

# Display the results
display(query_df)

# Close the connection
conn.close()
```

```python
# Pandas dataframe via sqlalchemy
conn = sqlalchemy.create_engine('sqlite:///../data/northwind.db') # ensure this is the correct path for the sqlite file. 
query_df = pd.read_sql_table("Order Details",conn) # retrieve the whole table and store in df

query_df = query_df[['ProductID','Quantity']].groupby('ProductID').sum().sort_values('Quantity',ascending=False)
query_df = query_df.reset_index(drop=False)
display(query_df)
```
